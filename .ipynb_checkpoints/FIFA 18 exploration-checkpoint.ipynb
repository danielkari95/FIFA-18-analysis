```python
import numpy as np
import numpy.random as npr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
fifa18 = pd.read_csv("data/fifa18.csv", index_col=0, low_memory=False)
```

Before any kind of analysis is conducted, it is always highly recommended to look into data.

```python
fifa18.head()
```

Seems that the players are sorted by `Overall`.

```python
fifa18.describe()
```

```python
fifa18.info()
```

```python
print(fifa18['Acceleration'].max())
print(fifa18['Acceleration'].min())
```

As we can notice, player attribute columns (`Acceleration` to `Volleys`) are of type `object`. We want to convert the attributes to numeric, since in reality each attribute gets integer value from 1 to 99. Player ID is also a categorical variable. 

Columns from `CAM` to `ST` excluding `ID` and `Preferred Positions` indicate what is the in-game overall of the player when playing on that position. These columns contain lots of missing values, because for goalkeepers these values are not present. This is important observation, because if we were to drop all the rows with missing values, we would accidentally get rid of all the goalkeepers. 

```python
fifa18[fifa18['Preferred Positions'].str.contains('GK')].info()
```

```python
fifa18.loc[:,'Acceleration':'Volleys'] = fifa18.loc[:,'Acceleration':'Volleys'].apply(pd.to_numeric, 
                                                                                      errors='coerce')
fifa18['ID'] = fifa18['ID'].astype('category')
```

```python
fifa18.info()
```

Conversion has generated some missing values, but that is not a big issue. For further analysis, we can drop the players who have missing attributes. 

```python
fifa18 = fifa18.dropna(subset=fifa18.loc[:,:'Volleys'].columns)
```

Now lets examine the data visually. How are the player overall values distributed?

```python
bins = len(fifa18['Overall'].unique())

sns.set()
fig, ax = plt.subplots(figsize=(16, 8))

sns.distplot(fifa18['Overall'], bins=bins)

plt.xlabel("Player overall", fontsize=20)
plt.show()
```

```python
from scipy.stats import norm

fig, ax = plt.subplots(figsize=(16, 8))

sns.distplot(fifa18['Overall'], bins=bins, kde=False, fit=norm)

plt.xlabel("Player overall", fontsize=20)
plt.show()
```

The column `Overall` seems to be approximately normally distributed. This is useful for futher analysis.

Let's look at the relationship between the attributes. Simply for the sake of plot clarity, only first 5 attributes and 100 random players are selected. 

```python
attributes = fifa18.loc[:, "Acceleration":"Volleys"]
```

```python
npr.seed(13)

sns.pairplot(attributes.iloc[npr.choice(len(attributes),size=100), :5])

plt.show()
```

We can also create a correlation matrix to inspect the relationships between the attributes.

```python
corr_matrix = attributes.corr(method='pearson')
corr_matrix.style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

There seems to be strong positive correlation e.g. between `Sprint speed` and `Acceleration` and between `Ball control` and `Dribbling`. On the other hand `Strength` does not seem to correlate with any other attribute. 

We can also see that the goalkeeper attributes correlate with each other but not with any other attributes.

```python
attributes[fifa18['Preferred Positions'].str.startswith('GK')].describe()
```

What about the correlation between the attributes and `Overall`?

```python
corrs = attributes.corrwith(fifa18['Overall']).to_frame()
corrs.style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=0)
```

One interesting question is if we know value for each attribute, can we predict the player overall rating. In order to study this properly, it is considered good practice to first split the data to training and test set.  

```python
npr.seed(13)

index_train = npr.choice(len(fifa18), size=2000, replace=False) # Randomly select training set of size 2000

fifa18_train = fifa18.iloc[index_train, :].loc[:, :'Volleys'] # Select the training data and drop columns after attributes since we don't really need them
```

Now we are ready to fit the model.

```python
X_train = fifa18_train.loc[:, 'Acceleration':'Volleys'] # The attributes are the explanatory variables
y_train = fifa18_train["Overall"] # Overall rating is the response we are trying to predict
```

```python
X_train = sm.add_constant(X_train) # Add constant term (intercept) to model.
model = sm.OLS(y_train, X_train) 
result = model.fit()

print(result.summary())
```

The $R^2$ might seem low, but there are couple of good reasons for this. It is known fact that overall is calculated differently for players in different positions. This means that we could get higher $R^2$-value if we fit the model e. g. only for strikers ('ST'). In addition to that, in the game each player also has ratings for skill moves and weaker foot (rated 1-5 stars) and attacking and defensive working rate (Low/Medium/High). These attributes might also affect the overall, but unfortunately they are not present in our dataset. 

It is also noteworthy that not all the attributes seem to be statistically significant predictors and some even seem to have negative effect. This is likely be caused by multicollinearity between the explanatory variables. Variance inflation factor (VIF) is one commonly used method to detect multicollinearity.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

for i in range(1, len(X_train.columns)):
    print("VIF for {0}: {1}".format(X_train.columns[i], vif(X_train.values, i)))
```

Luckily multicollinearity is not that serious problem since it doesn't affect prediction, which we are interested in. 

The next step is to use the remaining data as the test set.

```python
fifa18_test = fifa18[fifa18.index.isin(index_train)==False].loc[:, :'Volleys'] # Select only the rows that are not in the training set

X_test = fifa18_test.loc[:, 'Acceleration':'Volleys'] 
y_test = fifa18_test["Overall"]
```

Now it's time to find out how our model performs with the test data. Let's use mean squared error as the performance metric. 

```python
from sklearn.metrics import mean_squared_error

print("Mean squared error for the test data:", mean_squared_error(result.predict(sm.add_constant(X_test)), y_test))
print("Mean squared error for the train data:", mean_squared_error(result.fittedvalues, y_train))
```

From statistical point of view, it is also crucial to check that the assumptions of linear model are valid. Probably the most fundamental assumption is that the error terms are independent and normally distributed with zero mean and equal variance. Error terms are defined as the difference between the actual response value and the predicted (fitted) value. Common problem for regression models is heteroscedasticity, which means that the error terms typically increase with the predicted values.

```python
fig, ax = plt.subplots(figsize=(16,8))

sns.scatterplot(result.fittedvalues, result.resid)

plt.ylabel("Residuals", fontsize=20)
plt.xlabel("Fitted values", fontsize=20)
plt.show()
```

The scatterplot looks pretty random and supportive of the assumption of equal variance. We can see that there are straight lateral lines in the plot, but this is most likely caused by the fact the attributes are integers and thus only have limited number of possible values. 

Let's also inspect the normality assumption of the error terms. 

```python
from scipy.stats import probplot

sns.set()

fig, ax = plt.subplots(figsize=(16,8))

plot = probplot(result.resid, plot=plt)

plt.show()
```

The error terms seem to be distributed as they ideally should. 

Another interesting question about the attributes is whether we could predict player's position based on his individual attributes. Regarding this it seems almost trivial task to distinct outfield players from goalkeepers, since the latter have their own set of attributes which mostly determine the overall. In the game forwards tend to be good at shooting, whereas midfielders are generally good passers and defenders naturally have high defensive attributes.

```python
def position(preferred_positions):
    if preferred_positions.startswith(('ST','CF','LW','LF','RF','RW')):
        return "Forward"
    elif preferred_positions.startswith(('CM','CAM','CDM','LM','RM')):
        return "Midfielder"
    else:
        return "Defender"
```

```python
outfield_players = fifa18[~fifa18.loc[:,'Preferred Positions'].str.startswith(('GK'))]

outfield_players.loc[:,'Position'] = outfield_players.loc[:,'Preferred Positions'].apply(position)
```

```python
from sklearn.model_selection import train_test_split

X_train_of, X_test_of, y_train_of, y_test_of = train_test_split(outfield_players.loc[:, attributes.columns], 
                                                                outfield_players['Position'], random_state=13)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
pca.fit(X_train_of)
```

```python
print("Explained variance per component:", pca.explained_variance_ratio_) 
print("Total explained variance:", np.sum(pca.explained_variance_ratio_))
```

```python
X_train_of_transform = pca.fit_transform(X_train_of)
X_test_of_transform = pca.transform(X_test_of)
```

```python
plt.style.use('default')
fig, ax = plt.subplots(figsize=(16, 8))
sns.scatterplot(x=X_train_of_transform[:,0], y=X_train_of_transform[:,1], hue=y_train_of)

plt.xlabel("Component 1", fontsize=20)
plt.ylabel("Component 2", fontsize=20)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
```

```python
knn_clf = knn(n_neighbors=5)

knn_clf.fit(X_train_of_transform, y_train_of)

print(knn_clf.score(X_train_of_transform, y_train_of))
```

```python
ks = np.arange(1, 20)

for k in ks:
    knn_clf = knn(n_neighbors=k)
    knn_clf.fit(X_train_of_transform, y_train_of)
    print("k={0}:".format(k), np.around(knn_clf.score(X_test_of_transform, y_test_of), decimals=3))
```

What if we use non-decomposed attributes instead? 

```python
for k in ks:
    knn_clf = knn(n_neighbors=k)
    knn_clf.fit(X_train_of, y_train_of)
    print("k={0}:".format(k), np.around(knn_clf.score(X_test_of, y_test_of), decimals=3))
```

When PCA is used to transform the attributes it seems that the prediction accuracy is only 5 percent lower. 

How well do other classifiers perform? Let's use logistic regression and naive Bayes and compare the results. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB 
```

```python
clf_logit = LogisticRegression(random_state=13, solver='lbfgs', max_iter=500, multi_class='multinomial').fit(X_train_of, y_train_of)
print("Mean accuracy:", clf_logit.score(X_test_of, y_test_of)) 
```

```python
clf_nb = MultinomialNB().fit(X_train_of, y_train_of)
print("Mean accuracy:", clf_nb.score(X_test_of, y_test_of)) 
```
